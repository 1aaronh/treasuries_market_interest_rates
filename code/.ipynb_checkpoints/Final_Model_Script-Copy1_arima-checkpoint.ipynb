```python
train_sequences = TimeseriesGenerator(Xs_train,y_train,length=53,batch_size=64)
test_sequences = TimeseriesGenerator(Xs_test,y_test,length=53,batch_size=64)
input_shape = train_sequences[0][0][0].shape
```

```python
rnn = Sequential()
rnn.add(GRU(16, input_shape=input_shape, return_sequences=True))
rnn.add(GRU(8, return_sequences=False))
# leave default activation for GRUs
rnn.add(Dense(4, activation='relu'))

rnn.add(Dense(1, activation='linear'))
```

```python
rnn.compile(optimizer=Adam(lr=.0005), loss='mape',metrics=['mse'])

history = rnn.fit(train_sequences, validation_data=test_sequences, epochs=10)
```

```python
plt.plot(history.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'], label='Test loss')
plt.legend();
```

```python
plt.plot(history.history['mse'], label='Train mse')
plt.plot(history.history['val_mse'], label='Test mse')
plt.legend();
```

```python
# use dropout

from tensorflow.keras.callbacks import EarlyStopping

rnn2 = Sequential()
rnn2.add(GRU(16, input_shape=input_shape, return_sequences=True)) # True if next layer is RNN
rnn2.add(GRU(8, return_sequences=False)) # False if nex
# leave default activation layer for GRUs
rnn2.add(Dense(4, activation='relu'))

rnn2.add(Dense(1, activation='linear'))

rnn2.compile(optimizer='adam', loss='mape', metrics=['mse'])

es = EarlyStopping(patience = 5)

history2 = rnn2.fit(train_sequences, validation_data=test_sequences, epochs=15, batch_size = 64, callbacks = [es])
```

```python
# https://stackoverflow.com/questions/57002948/how-to-calculate-mean-absolute-percentage-error-for-regression-probem-in-python

def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
# Code modified from code written by Matthew Garton.

def plot_series(df, cols=None, title='Title', xlab=None, ylab=None, steps=1):
    
    # Set figure size to be (18, 9).
    plt.figure(figsize=(18,9))
    legend_list = []
    
    # Iterate through each column name.
    for col in cols:
            
        # Generate a line plot of the column name.
        # You only have to specify Y, since our
        # index will be a datetime index.
        plt.plot(df[col])
        legend_list.append(col)
        
    # Generate title and labels.
    plt.title(title, fontsize=26)
    plt.xlabel(xlab, fontsize=20)
    plt.ylabel(ylab, fontsize=20)
    
    # Enlarge tick marks.
    plt.yticks(fontsize=18)
    plt.xticks(df.index[0::steps], fontsize=18)
    
    plt.legend(legend_list);
```

```python
# Code written by Joseph Nelson.
# Improved by Hovanes Gasparian

def interpret_dftest(dftest):
    dfoutput = pd.Series(dftest[0:3], index=['Test Statistic','p-value', 'Lag Used'])
    return dfoutput
```

```python
# https://www.machinelearningplus.com/time-series/vector-autoregression-examples-python/
# use with forecast_var2 and dfvar
def invert_transformation(df_train, df_forecast, second_diff=False):
    """Revert back the differencing to get the forecast to original scale."""
    df_fc = df_forecast.copy()
    columns = df_train.columns
    for col in columns:        
        # Roll back 1st Diff
        df_fc[str(col)+'_forecast'] = df_train[col].iloc[-1] + df_fc[str(col)+'_1d'].cumsum()
# set to new dataframe
    return df_fc

# this needs to work for plotting and metrics
```

```python
# https://www.machinelearningplus.com/time-series/vector-autoregression-examples-python/
def forecast_accuracy_MAPE(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    print(f'MAPE: {mape}')
```
