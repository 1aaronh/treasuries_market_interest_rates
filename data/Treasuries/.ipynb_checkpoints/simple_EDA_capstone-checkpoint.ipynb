```python
def DateFinder(string):
    string = string[7:]
    return string

def InterestFinder(string):
    string = string[3:6]
    return string
```

```python
string = '04.875 06/30/12'
DateFinder(string)
```

```python
InterestFinder(string)
```

```python
df['interest_rate'] = df['Security description'].map(InterestFinder)
df['maturity_date'] = df['Security description'].map(DateFinder)
```
