```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GRU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# modeldata.to_csv('../DataBank/csv_exports/rnn_nocp.csv')
```

```python
modelcsv = pd.read_csv('../DataBank/csv_exports/rnn_nocp.csv')
```

```python
modelcsv.head(1)
```

```python
modelcsv = modelcsv.rename(columns={'Unnamed: 0':'Date'})


modelcsv['Date'] = pd.to_datetime(modelcsv['Date'])
modelcsv.index = modelcsv['Date']

modelcsv = modelcsv.drop(columns=['unique_date','Settlement date','Trade date',
                                  'Trade amount (in millions, USD)','Term','Date'])

features = [col for col in modelcsv.columns if col != '30Yr-1MO Spread']
X = modelcsv[features]
y = modelcsv['30Yr-1MO Spread']
```

```python
X.head(1)
```

```python
X.index.name = None
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,shuffle=False)
```

```python
X.dtypes
```

```python
# StandardScaler cannot accept datetime datatypes

ss = StandardScaler()
Xs_train = ss.fit_transform(X_train)
Xs_test = ss.transform(X_test)
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state = 42,n_components=10)
pca.fit(Xs_train)
Z_train = pca.transform(Xs_train)
Z_test = pca.transform(Xs_test)
```

```python
# PCA RNN model

train_sequences = TimeseriesGenerator(Z_train,y_train,length=10,batch_size=64)
test_sequences = TimeseriesGenerator(Z_test,y_test,length=10,batch_size=64)
input_shapeZ = train_sequences[0][0][0].shape
```

```python
train_sequences = TimeseriesGenerator(Xs_train,y_train,length=3,batch_size=64)

test_sequences = TimeseriesGenerator(Xs_test,y_test,length=3,batch_size=64)
```

```python
train_sequences[0][0].shape
```

```python
input_shape = train_sequences[0][0][0].shape
```

```python
input_shape
```

```python
model = Sequential()
model.add(GRU(8, input_shape=input_shape, return_sequences=True)) # True if next layer is RNN
model.add(GRU(8, return_sequences=False)) # False if next layer is Dense

model.add(Dense(4, activation='relu'))

model.add(Dense(1, activation='linear'))
```

```python
model.compile(optimizer=Adam(lr=.0005), loss='mse', metrics=['mse'])

history = model.fit(train_sequences, validation_data=test_sequences, epochs=10)
```

```python
plt.plot(history.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'], label='Test loss')
plt.legend();
```
