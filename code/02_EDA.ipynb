### Exploratory Data Analysis

```python
# Imports: Visualization and Data Processing

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Ease for display
sns.set_style(style="whitegrid")
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('../data/csv_exports/edaFINAL.csv')
```

```python
df.head(3)
```

```python
# adjust csv import
df.drop(columns=['Unnamed: 0','index'],inplace=True)
```

```python
df['Trade date'] = pd.to_datetime(df['Trade date'])
```

```python
df.index = df['Trade date']
```

```python
# Make new DataFrames for vsualizing purchase and sale categories
# This relates to whether Primary Dealer counterparties were buying or selling Treasuries at the given transactions
# Generally, many more transactions were NYFed Buying (Counterparties selling) Treasuries as part of QE program

dfsale = df[df['Transaction category'] == 'Sale'].copy()
dfpurchase = df[df['Transaction category'] != 'Sale'].copy()
```

```python
# year or month can be adjusted with string value
# Counterparties with most transactions (NYFed selling)
# remove.loc for entire time series

year = '2011'
print(f'{year}: Top 15 Buyers from NYFed')
dfsale.loc[year].groupby('Counterparty')['Transaction category'].count().sort_values(ascending=False).head(10)
```

```python
# Counterparties with most transactions (NYFed buying)

year = '2014'
print(f'{year}: Top 15 Sellers to NYFed')
print
dfpurchase.loc[year].groupby('Counterparty')['Transaction category'].count().sort_values(ascending=False).head(15)
```

```python
# same as above: as bar plot

year = '2014'
# print(f'{year}: Top 15 Sellers to NYFed')

plt.figure(figsize=(12,10))
dfpurchase.loc[year].groupby('Counterparty')['Transaction category'].count().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title(f'{year} Top 10 Total Sellers to NYFed',size=16)
plt.ylabel(None)
plt.xlabel('Transactions for Year',size=14)
plt.yticks(size=12);
```

```python
# Descriptive statistics by Counterparty on avg time to matruity (portfolio exposure sslected)
# NYFed Sells - Counterparty exposure added
# year can be made variable

dfsale.loc['2012'].groupby('Counterparty')['Time to Maturity (Years)'].mean().head(10)
```

```python
# Descriptive statistics by Counterparty on avg time to matruity (portfolio exposure sslected)
# NYFed Buys - Counterparty exposure subtracted
# year can be made variable

dfpurchase.loc['2012'].groupby('Counterparty')['Time to Maturity (Years)'].mean().head(10)
```

```python
# barplot: same as above

dfpurchase.loc['2013'].groupby('Counterparty')['Time to Maturity (Years)'].mean().head(10).plot(kind='bar')
```

## Scatterplots for Exploring Character of Transactions by Counterparty

```python
# Template for selecting from Top 10 Counterparties

dfplot1 = df[df['Counterparty'] == 'BMO Capital Markets Corp.']
dfplot1 = dfplot1.loc['2011']

plt.figure(figsize=(10,8))
sns.scatterplot(x=dfplot1['Time to Maturity (Years)'],
                y=dfplot1['Coupon'],
               hue=dfplot1['Type'].tolist(),
               data=dfplot1)
```

```python
# number of counterparties to consider
len(list(df['Counterparty'].value_counts().keys()))
```

```python
# whole time series: Types of treasuries Party purchased by remaining maturity and Coupon (quoted interest rate) 

# first 10 counterparties
# central bank sales to them
# only years 2011 and 2012 have data for NY Fed selling Treasuries

# select names of counterparties
party_list = list(df['Counterparty'].value_counts().keys())
party_list = party_list[:11]

# toggle year 2010-2014
# some years do not have data for certain counterparties
year = '2011'

plt.figure(figsize=(10,8))
for party in party_list:

    dfplot1 = dfsale[dfsale['Counterparty'] == party]
    dfplot1 = dfplot1.loc[year]
    
    # plot transactions
    plt.figure(figsize=(8,6))
    plt.title(f'{year} Purchases from NYFed: {party}')
    sns.scatterplot(x=dfplot1['Time to Maturity (Years)'],
                    y=dfplot1['Coupon'],
                   hue=dfplot1['Type'].tolist(),
                   data=dfplot1)
```

```python
# second 10 counterparties sales to them from NYFed:
# only years 2011 and 2012 have data for NY Fed selling Treasuries

party_list = list(df['Counterparty'].value_counts().keys())
party_list = party_list[11:21]
year = '2011'

for party in party_list:

    dfplot1 = dfsale[dfsale['Counterparty'] == party]
    
    # 2010-2014 loc ranges
    dfplot1 = dfplot1.loc[year]
    
    # plot transactions
    plt.figure(figsize=(8,6))
    plt.title(f'{year} Purchases from NYFed: {party}')
    sns.scatterplot(x=dfplot1['Time to Maturity (Years)'],
                    y=dfplot1['Coupon'],
                   hue=dfplot1['Type'].tolist(),
                   data=dfplot1)
```

## Rolling Averages for Top 10 Counterparties over Time Series (Purchase & Sale)

```python
# Test for rolling average plot
df[df['Counterparty'] == 'Morgan Stanley & Co. LLC']['Time to Maturity (Years)'].rolling(1000).mean().plot()
```

```python
# Test for rolling average plot
df[df['Counterparty'] == 'Morgan Stanley & Co. LLC']['Coupon'].rolling(1000).mean().plot()
```

```python
# Top 5 counterparties
# central bank purchases from them
# Rolling Mean of Coupon (interest paid on Treasuries)

tops = list(dfpurchase.groupby('Counterparty')['Transaction category'].count().sort_values(ascending=False).keys())
top5 = tops[:5]

# https://stackoverflow.com/questions/35164019/filter-multiple-values-using-pandas
# Filter by top 10 counterparty list
# dfplot1 = dfpurchase[dfpurchase['Counterparty'].isin(top10)]

# https://stackoverflow.com/questions/26355313/plotting-multiple-plots-generated-inside-a-for-loop-on-the-same-axes-python
# for loop solution on one plot

measure = 'Coupon'
plt.figure(figsize=(18,14))
plt.title(f'Top 5 Sellers to NYFed by {measure} (Rolling Average)',size=20)

for party in top5:

    dfplot1 = dfpurchase[dfpurchase['Counterparty'] == party]
    sns.lineplot(x=dfplot1['Trade date'],
                    y=dfplot1[measure].rolling(1500).mean(),
                   data=dfplot1,label=party)
    
plt.ylabel(measure,size=16)
plt.xlabel('Date',size=16)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(prop={'size':18});
```

```python
# sorted method from here
# https://www.kite.com/python/answers/how-to-sort-a-list-alphabetically-in-python

# Primary Chart with this method

# Top 5 counterparties
# central bank purchases from them
# Rolling Mean of Time to Maturity 

tops = list(dfpurchase.groupby('Counterparty')['Transaction category'].count().sort_values(ascending=False).keys())
top5 = tops[:5]
top5 = sorted(top5)
print(top5)

measure = 'Time to Maturity (Years)'
plt.figure(figsize=(18,14))
plt.title(f'Top 5 Sellers to NYFed by {measure} (Rolling Average)',size=20)

for party in top5:
    
    dfplot1 = dfpurchase[dfpurchase['Counterparty'] == party]
    sns.lineplot(x=dfplot1['Trade date'],
                    y=dfplot1[measure].rolling(1500).mean(),
                   data=dfplot1,label=party)
    
plt.ylabel(measure,size=16)
plt.xlabel('Date',size=16)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(prop={'size':18});
```
